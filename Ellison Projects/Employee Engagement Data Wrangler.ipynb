```python
import os
import pandas as pd
from tqdm import tqdm
        
files = [file for file in os.listdir() if (file[-4:]=='xlsx') & (file!='Aggregate Data.xlsx')]
questions = ['Treatment by local operation', 'Trust in local operation',
       'Enjoyment in day-to-day tasks', 'Look forward to work',
       'Confidence in future of local operation',
       'Confidence in career trajectory in operation',
       'Helping others with heavy workloads', 'Extra effort in job',
       'Looking for jobs elsewhere', 'Sent out phone calls/resume elsewhere',
       'Belief in organization\'s values', 'Value of employee feedback',
       'Local operation encourages work-life balance',
       'Local team cooperation',
       'Manager communication (clear feedback on work/performance)',
       'Manager helps me understand responsibilities',
       'Adequate resources to fulfill responsibilities',
       'Job utilizes skills and abilities well',
       'Understand career opportunities at organization',
       'Understands contribution to organization', 'Fair compensation',
       'Satisfied with frequency of individual performance/work recognition',
       'Confidence in senior leadership decisions in local operation',
       'Understand contribution to end customer satisfaction']
combined_data = pd.DataFrame(columns=questions)

for file in tqdm(files, position=0, leave=True):
    temp_data = pd.DataFrame(columns=questions)
    totals_helper = pd.read_excel(file, sheet_name=list(pd.read_excel(file, sheet_name=None).keys())[0])
    totals = int(totals_helper['Unnamed: 2'][7] + totals_helper['Unnamed: 2'][8])
    temp_data['Office'] = [file[:-5]]*totals
    for sheet in pd.read_excel(file, sheet_name=None).keys():
        helper_df = pd.read_excel(file, sheet_name=sheet)
        helper_df = helper_df.dropna().reset_index(drop=True)
        helper_df['Unnamed: 2'] = [int(value) if value>-1 else value for value in helper_df['Unnamed: 2']]
        answers = ([2]*helper_df['Unnamed: 2'][0]) + ([1]*helper_df['Unnamed: 2'][1]) + ([0]*helper_df['Unnamed: 2'][2]) + ([-1]*helper_df['Unnamed: 2'][3]) + ([-2]*helper_df['Unnamed: 2'][4]) + (['Skip']*(totals-sum(helper_df['Unnamed: 2'][:5])))
        temp_data[temp_data.columns[int(sheet[-2:])-1]] = answers
    combined_data = combined_data.append(temp_data).reset_index(drop=True)
        
combined_data[[combined_data.columns[8], combined_data.columns[9]]] = combined_data[[combined_data.columns[8], combined_data.columns[9]]].multiply(-1, axis="index")
combined_data
```

```python
combined_data.to_excel('Aggregate Data.xlsx')
```
