```python
from scipy import stats
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from tqdm import tqdm

redvgreen = pd.read_excel('Market Share Analysis.xlsx', sheet_name='Red v Green TTM')

redvgreen = redvgreen[7:]
redvgreen = redvgreen.drop(redvgreen.columns[:5], 1)
redvgreen = redvgreen.drop(redvgreen.columns[10:], 1)
redvgreen = redvgreen.dropna()
redvgreen = redvgreen.reset_index(drop=True)

for index, row in redvgreen.iterrows():
    if type(row['Unnamed: 6']) == str:
        redvgreen = redvgreen.drop([index])

redvgreen = redvgreen.reset_index(drop=True)
redvgreen
```

```python
mktshare_pvalues = []

for index,row in tqdm(redvgreen.iterrows()):
    FY2019 = [1] * row[3] + [0] * (row[4] - row[3])
    FY2020 = [1] * row[0] + [0] * (row[1] - row[0])
    if np.mean(FY2019) > np.mean(FY2020):
        mktshare_pvalues.append(stats.ttest_ind(FY2019, FY2020, equal_var=False, alternative='greater').pvalue)
    else:
        mktshare_pvalues.append(stats.ttest_ind(FY2019, FY2020, equal_var=False, alternative='less').pvalue)

redvgreen['mktshare_pvalues'] = mktshare_pvalues

redvgreen.to_excel('redvgreen.xlsx')
```

```python
allmachines = pd.read_excel('Market Share Analysis.xlsx', sheet_name='All Machines')
allmachines = allmachines[allmachines.columns[2:12]]
allmachines = allmachines.dropna()
allmachines = allmachines.reset_index(drop=True)

for index, row in allmachines.iterrows():
    if type(row['Unnamed: 2']) == str:
        allmachines = allmachines.drop([index])

allmachines = allmachines.astype(int)
allmachines
```
